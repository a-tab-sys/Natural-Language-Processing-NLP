```python
# importing libraries
from tensorflow.keras.layers import Embedding  #helps us with word2vec implemetation
from tensorflow.keras.preprocessing.sequence import pad_sequences  #used for padding (pre or post)
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
### Vocabulary size - num of unique words
voc_size=5000
```

### Onehot Representation

```python
messages=X.copy()
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
# 1. import dataset
# 2. independant and dependant features
# 3. cleaning the data 1)stemming, 2)stopwords
# 4. fix a sentance length to fix the input
# 5. one hot representation
# 6. create embedding layer
# 7. LSTM neural network
```

```python
nltk.download('stopwords')
```

```python
print(messages.index)
```

```python
print(messages.isnull().sum())
```

```python
### Dataset Preprocessing
from nltk.stem.porter import PorterStemmer ##stemming purpose
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    # review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])   # gives error because index 6 does not exist
    # Use .iloc[] for positional indexing: If you're trying to access rows by their position (rather than their index label), you can use .iloc[] instead of []. This will access the row by its integer position rather than by the index value.For example:
    review = re.sub('[^a-zA-Z]', ' ', messages['title'].iloc[i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)

# could have also applied lemmatization but this will take more time
```

```python
corpus
```

```python
corpus[1]
```

```python
onehot_repr=[one_hot(words,voc_size)for words in corpus]
onehot_repr
```

```python
corpus[1]
```

```python
onehot_repr[1]
```

### Embedding Representation

```python
sent_length=20
# selected 20 because mos sentances are between 10-12 words. 20 should be enough to cover all the sentances
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[1]
```

```python
embedded_docs[0]
```

```python
# why not tfidf?? 

#because with lstm we use a word emedding layer. tfidf is not that effeciant compared to word2vec and word embedding
```

```python
## Creating model
embedding_vector_features=40 ##features representation # this means that each sentance which is represented by 20 numbers, each of those 20 numbers will be represented by a vector of length 40
# changing this value is called hyper parameter tuning
model=Sequential()  #use sequencial model
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))  #creating embedding layer
model.add(LSTM(100))  #ltms ith 100 neurons - could be 200, 300 up to you
model.add(Dense(1,activation='sigmoid'))  #my output is binary, so we use sigmoid activation function
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])  #loss=binary_crossentropy because we have binary output
print(model.summary())
```

```python
len(embedded_docs),y.shape
```

```python
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

### Model Training

```python
### Finally Training
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

### Adding Dropout

```python
from tensorflow.keras.layers import Dropout
## Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

### Performance Metrics And Accuracy

```python
y_pred=model.predict(X_test)
```

```python
y_pred=np.where(y_pred > 0.6, 1,0) ##AUC ROC Curve #if you dont want threshold of 0.5 use auc roc curve
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
